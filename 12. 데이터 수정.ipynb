# 데이터 수정

```python
import pandas as pd
df=pd.read_excel("score.xlsx", index_col="지원번호")
df
```

## Column 수정

```python
df['학교'].replace({'북산고': '상북고', '능남고': '무슨고'})
```

```python
df
```

```python
df['학교'] = df['학교'].replace({'북산고': '상북고'})
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기']=df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교']+'등학교'
```

```python
df
```

Column 추가

```python
df['총합']=df['국어']+df['영어']+df['수학'] +df['과학'] +df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 컬럼을 만들고 기본값으로 Fail을 넣어준다.
df
```

```python
df.loc[df['총합']>400, '결과'] = 'Pass' # 총합이 400점 이상인 경우 결과를 Pass로 바꿔준다.
df
```
