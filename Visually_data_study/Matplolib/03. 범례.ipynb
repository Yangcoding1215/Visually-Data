# 03. 범례

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1 , 2, 3]
y=[2, 4, 8]
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='upper right')
# loc='best' : 자동으로 가장 적절한 위치
# loc='upper right' : 오른쪽 위
# loc='upper left' : 왼쪽 위
# loc='lower right' : 오른쪽 아래
# loc='lower left' : 왼쪽 아래
# loc='right' : 오른쪽 가운데
# loc='center left' : 왼쪽 가운데
# loc='center right' : 오른쪽 가운데
# loc='lower center' : 아래 가운데
# loc='upper center' : 위 가운데
# loc='center' : 정가운데
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.7, 0.8)) # x축, y축 (0~1사이)
```
