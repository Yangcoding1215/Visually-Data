## 선 스타일

```python
plt.plot(x,y, linestyle=':') #점선
```

```python
plt.plot(x,y, linestyle='--') #점선
```

```python
plt.plot(x,y, linestyle='--') # 긴 점선
```

```python
plt.plot(x,y, linestyle='-.') #점선과 실선을 번갈아가며 그리는 스타일
```

```python
# Matplotlib 라인 스타일 전체 정리
#
# 1. 기본 사용법
#    plt.plot(x, y, linestyle='--')
#    plt.plot(x, y, ls='--')        # linestyle의 축약형은 ls
#
# 2. 가장 많이 쓰는 기본 라인 스타일
#    '-' 또는 'solid'
#      : 실선
#      : 가장 기본적인 선. 데이터 흐름을 자연스럽게 보여줄 때 사용.
#
#    '--' 또는 'dashed'
#      : 파선(점선처럼 보이는 긴 끊긴 선)
#      : 기준선, 비교선, 보조선 표현에 자주 사용.
#
#    ':' 또는 'dotted'
#      : 점선
#      : 약한 강조, 참고용 선, 보조 정보 표시할 때 적합.
#
#    '-.' 또는 'dashdot'
#      : 일점쇄선(선-점-선-점)
#      : 다른 선과 확실히 구분하고 싶을 때 유용.
#
# 3. 선을 아예 그리지 않는 경우
#    'None', 'none', ' ', ''
#      : 선 없음
#      : 마커(marker)만 표시하고 싶을 때 사용.
#
#    예시
#    plt.plot(x, y, linestyle='None', marker='o')
#
# 4. 문자열 별칭 정리
#    '-'        == 'solid'
#    '--'       == 'dashed'
#    ':'        == 'dotted'
#    '-.'       == 'dashdot'
#    'None'     == 'none' == ' ' == ''
#
# 5. 사용자 정의 대시 패턴(custom dash pattern)
#    튜플 형태로 세부 패턴을 직접 지정할 수 있다.
#    형식: (offset, (on_off_seq))
#
#    offset
#      : 대시 패턴이 시작되는 위치
#
#    on_off_seq
#      : 선을 그리는 길이와 비우는 길이를 번갈아 지정
#      : 단위는 포인트(point)
#
#    예시
#    plt.plot(x, y, linestyle=(0, (5, 2)))
#      : 5pt 그린 뒤 2pt 비우기 반복
#
#    plt.plot(x, y, linestyle=(0, (3, 1, 1, 1)))
#      : 3pt 선, 1pt 공백, 1pt 선, 1pt 공백 반복
#
# 6. 자주 같이 쓰는 옵션
#    linewidth 또는 lw
#      : 선 두께 지정
#      : 예) plt.plot(x, y, linestyle='--', linewidth=3)
#
#    color 또는 c
#      : 선 색상 지정
#      : 예) plt.plot(x, y, linestyle=':', color='red')
#
#    marker
#      : 데이터 점 모양 지정
#      : 예) plt.plot(x, y, linestyle='-.', marker='o')
#
#    alpha
#      : 투명도 지정(0~1)
#      : 예) plt.plot(x, y, linestyle='--', alpha=0.5)
#
# 7. 참고하면 좋은 조합 예시
#    plt.plot(x, y, ls='-')
#      : 기본 데이터 선
#
#    plt.plot(x, y, ls='--', lw=2)
#      : 비교용 보조 선
#
#    plt.plot(x, y, ls=':', c='gray')
#      : 약하게 보이는 참고 선
#
#    plt.plot(x, y, ls='-.', marker='o')
#      : 선과 점을 함께 강조
#
#    plt.plot(x, y, ls='None', marker='s')
#      : 선 없이 데이터 점만 표현
#
# 8. 한 줄 요약
#    실선(-), 파선(--), 점선(:), 일점쇄선(-.)이 기본이고,
#    필요하면 None으로 선을 없애거나 튜플로 대시 패턴을 직접 만들 수 있다.
```

##  축약어

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

```python
# matplotlib 자주 쓰는 축약어 정리
#
# 1. 선(line) 관련
# ls   = linestyle
#      : 선 스타일
#      : '-', '--', ':', '-.' 등을 지정할 때 사용
#      : 예) plt.plot(x, y, ls='--')
#
# lw   = linewidth
#      : 선 두께
#      : 예) plt.plot(x, y, lw=3)
#
# c    = color
#      : 선 또는 마커의 기본 색상
#      : 예) plt.plot(x, y, c='red')
#
# aa   = antialiased
#      : 선을 부드럽게 보이게 할지 여부
#      : 예) plt.plot(x, y, aa=True)
#
# 2. 마커(marker) 관련
# ms   = markersize
#      : 마커 크기
#      : 예) plt.plot(x, y, marker='o', ms=10)
#
# mec  = markeredgecolor
#      : 마커 테두리 색상
#      : 예) plt.plot(x, y, marker='o', mec='blue')
#
# mew  = markeredgewidth
#      : 마커 테두리 두께
#      : 예) plt.plot(x, y, marker='o', mew=2)
#
# mfc  = markerfacecolor
#      : 마커 내부 색상
#      : 예) plt.plot(x, y, marker='o', mfc='red')
#
# 3. 투명도, 이름 등
# alpha = alpha
#       : 투명도
#       : 보통 축약형 없이 그대로 사용
#       : 예) plt.plot(x, y, alpha=0.5)
#
# label = label
#       : 범례에 표시할 이름
#       : 보통 그대로 사용
#       : 예) plt.plot(x, y, label='data1')
#
# marker = marker
#        : 마커 모양
#        : 보통 그대로 사용
#        : 예) plt.plot(x, y, marker='s')
#
# 4. 자주 보는 조합 예시
# plt.plot(x, y, ls='--', lw=2, c='green')
# plt.plot(x, y, marker='o', ms=8, mec='black', mfc='yellow')
# plt.plot(x, y, marker='^', ls=':', c='red', alpha=0.7)
#
# 5. 한 줄 요약
# ls, lw, c는 선 관련 축약어이고,
# ms, mec, mew, mfc는 마커 관련 축약어로 자주 사용된다.
```
