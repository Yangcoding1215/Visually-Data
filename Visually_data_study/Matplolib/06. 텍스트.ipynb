# 6. 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1 , 2, 3]
y=[2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, marker='o')
for idx, txt in enumerate(y) : # 각 점에 y값을 텍스트로 표시, idx는 점의 인덱스, txt는 y값
    plt.text(x[idx], y[idx], txt) # plt.text() 함수는 그래프의 특정 위치에 텍스트를 추가하는 함수입니다.
```

```python
plt.plot(x,y, marker='o')
for idx, txt in enumerate(x) : # 각 점에 y값을 텍스트로 표시, idx는 점의 인덱스, txt는 y값
    plt.text(x[idx], y[idx], txt) # plt.text() 함수는 그래프의 특정 위치에 텍스트를 추가하는 함수입니다.
```

```python
plt.plot(x,y, marker='o')
for idx, txt in enumerate(y) : # 각 점에 y값을 텍스트로 표시, idx는 점의 인덱스, txt는 y값
    plt.text(x[idx], y[idx] +0.3, txt, ha='center', color='blue') # plt.text() 함수는 그래프의 특정 위치에 텍스트를 추가하는 함수입니다.
```
