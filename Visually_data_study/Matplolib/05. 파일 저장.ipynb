```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1 , 2, 3]
y=[2, 4, 8]
```

```python
plt.plot(x,y)
plt.savefig('graph.png') #그래프를 이미지 파일로 저장 (파일명과 확장자 지정)
```

```python
plt.plot(x,y)
plt.savefig('graph.png', dpi=100) #그래프를 이미지 파일로 저장 (파일명과 확장자 지정, 해상도 설정)
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.savefig('graph_200.png')
```

```python
plt.figure(dpi=200) # 그래프의 해상도 설정 (dpi: dots per inch)
plt.plot(x,y)
plt.savefig('graph_200.png', dpi=100) # 그래프를 이미지로 저장하되 figure의 dpi와 다른 해상도로 저장 (파일명과 확장자 지정, 해상도 설정)
```
