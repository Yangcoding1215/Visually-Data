# 함수 적용

```python
import pandas as pd

df=pd.read_excel("score.xlsx", index_col="지원번호")
# `키` 열은 처음에 숫자형인지 확인
print('dtype of 키 before any changes ->', df['키'].dtype)
df
```

```python
# 키 뒤에 cm을 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

# 적용 전 dtype
print('before apply:', df['키'].dtype)

df['키'] = df['키'].apply(add_cm)

# 적용 후 dtype
print('after apply :', df['키'].dtype)
df
```

```python
# 3. 함수로 처리하면서 원본 유지하기

# 열 자체를 변형하지 않고 보기용 Series를 생성
formatted_heights = df['키'].apply(lambda h: f"{h}cm")
formatted_heights.head()

# 필요하다면 새 컬럼에 할당하되 기존 컬럼은 변경하지 않음
# df['키_보기용'] = formatted_heights
```

```python
def Capitalize(lang):
    if pd.notnull(lang): # NAN이 아닌지
        return lang.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(Capitalize)
df
```
