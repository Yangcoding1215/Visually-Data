# 09. 데이터 선택 (조건)

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 True, False로 반환
```

```python
filt = (df['키'] >= 185) 
df[filt]
```

```python
df['키'] >= 185
```

```python
df[~filt] # 키가 185 미만인 학생들만 출력
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 점수
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] # 키가 185 이상인 학생들의 이름, 수학, 과학 점수
```

## 다양한  조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고') ] # 키가 185 이상이면서 학교가 북산고인 학생들의 데이터
```

### |또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키가 170 미만이거나 200 초과인 학생들의 데이터]
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # 송씨 성을 가진 사람
df[filt]
```

```python
filt= df['이름'].str.contains('태') # 이름에 '태' 글자가 포함된 사람
df[filt]
```

```python
df[~filt] #이름에 '태' 글자가 포함되지 않은 사람
```

```python
langs = ["Python", "Java"]
filt = df['SW특기'].isin(langs) # SW특기가 python이거나 Java인 사람
df[filt]
```

```python
df
```

```python
# 대문자인 Python이 포함된 사람을 포함시키려면
langs = ["python", "java"]
filt = df['SW특기'].str.lower().isin(langs) # SW특기의 모든 글자를 소문자로 바꿔서 비교
df[filt]
```

```python
# Java가 포함된 사람을 포함시키려면
filt = df['SW특기'].str.contains('Java')
df[filt]
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java', na=True)
```
