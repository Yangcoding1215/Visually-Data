## 데이터 제외하기 Dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna() # 결측치가 있는 행 제거
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

- axis : index or columns  
- how : any or all

```python
df.dropna(axis='index', how='any') # NAN이 하나라도 있는 행 제거
```

```python
df.dropna(axis='columns') # NAN이 하나라도 있는 열 제거
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how= 'all') # 데이터 전체가 NAN 경우에만 Column 삭제
```
